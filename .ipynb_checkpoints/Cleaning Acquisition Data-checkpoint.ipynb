```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
aq_data_to_load = "aquisition_data.csv"
```

```python
# Read the school data file and store it in a Pandas DataFrame.
aq_data_df = pd.read_csv(aq_data_to_load)
aq_data_df
```

```python
# Determine if there are any missing values in the acquisition data.
aq_data_df.count()
```

```python
# Determine if there are any missing values in the acquisition data.
aq_data_df.isnull()
```

```python
aq_data_df.isnull().sum()
```

```python
#determine data type
aq_data_df.dtypes
```

```python
aq_df = aq_data_df.drop(['Co-borrower Credit Score at Origination', 'Mortgage Insurance Type'], axis=1)
```

```python
aq_df.isnull().sum()
```

```python
clean_df = aq_df.dropna()
```

```python
clean_df.count()
```

```python
clean_df
```

```python
#Rename the columns 
clean_df = clean_df.rename(columns={'Loan Identifier': 'loan identifier', 'Origination Channel ': 'origination channel', 'Seller Name': 'seller name', 'Original Interest Rate': 'original interest rate', 'Original UPB': 'original upb', 'Original Loan Term': 'original loan term', 'Origination Date': 'origination date', 'First Payment Date': 'first payment date', 'Original Loan-to-value (LTV)': 'ltv', 'Original Combined Loan-to-value (CLTV)': 'cltv', 'Number of Borrowers': 'number of borrowers', 'Original Debt to Income Ratio ': 'original debt to income ratio', 'Borrower Credit Score at Origination': 'borrower credit score at origination', 'First Time Home Buyer Indicator': 'first time home buyer indicator', 'Loan Purpose': 'loan purpose', 'Property Type': 'property type', 'Number of Units': 'number of units', 'Occupancy Type': 'occupancy type', 'Property State': 'property state', 'Zip Code Short': 'zip code short', 'Product Type': 'product type', 'Relocation Mortgage Indicator': 'relocation mortgage indicator'})
```
