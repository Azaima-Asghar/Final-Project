##### *Remove the Servicer Name, Current Actual UPB, Zero Balance Code, Zero Balance Effective Date and Service Activity Indicator columns from the performance DataFrame.*

```python
# Drop the unwanted columns from the performance_data dataframe.
performance_df = performance_data.drop(['Servicer Name','Current Actual UPB','Zero Balance Code', 'Zero Balance Effective Date',
                                        'Service Activity Indicator'], axis = 1)
performance_df.head()
```

```python
# Create the clean dataframe with all null values removed.
clean_performance_df = performance_df.dropna()
clean_performance_df.head()
```

```python
# Count the number of values in each column.
clean_performance_df.count()
```

```python
# View the clean_performance_df.
clean_performance_df.head()
```

```python
# Rename the coulmns in the clean performance dataframe.
clean_performance_df = clean_performance_df.rename(columns = {'Loan Identifier':'loan identifier', 
'Monthly Reporting Period': 'monthly reporting period','Current Interest Rate':'current interest rate',
'Loan Age':'loan age','Remaining Months to Legal Maturity':'remaining months to legal maturity',
'Adjusted Months to Maturity':'adjusted months to maturity','Maturity Date':'maturity date',
'Metropolitan Statistical Area (MSA)': 'MSA', 'Current Loan Delinquency Status':'current loan delinquency status',
'Modification Flag': 'modification flag'})
```

```python
# View the clean_performance_df after the names of the columns are changed.
clean_performance_df.head()
```

```python
#output_file = 'cleaned_Performance.csv'
output_file = 'cleaned_Performance_full.csv'
output_dir = pathlib.Path.cwd().parent / "CSV files" # can join path elements with / operator

# Save the cleaned performance dataframe into a csv file.
clean_performance_df.to_csv(output_dir / output_file, index = False)
```
